# Stock Prediction of Netflix

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
# Load the data
df = pd.read_csv('NFLX.csv')
df.head(6)
```

```python
# Get the number of trading days
df.shape
```

```python
# Visualize the close price data
plt.figure(figsize = (16, 8))
plt.title('Netflix')
plt.xlabel('Days')
plt.ylabel('Close price in USD ($)')
plt.plot(df['Close'])
plt.show()
```

```python
# Get the close price in a new Data Frame
df = df[['Close']]
df.head(4)
```

```python
# Create a variable to predict the x days out into the future
future_days = 25
# Create a new column that will contain our Target data to predict values for x future days
# This will be the shifted x days up
df['Prediction'] = df[['Close']].shift(-future_days)
print(df.head(4))
print()
print(df.tail(4))
```

```python
# Create the Featre Dataset (X) and convert it to a numpy array and remove the last x rows/days
# We drop the prediction column and we want all of the rows except for the last x days
X = np.array(df.drop(['Prediction'], 1))[:-future_days]
print(X)
```

```python
# Create the Target Dataset (y) and we will convert it in a numpy array
# We need all the Target values except the last x rows/days
y = np.array(df['Prediction'])[:-future_days]
print(y)
```

```python
# Split the Data into 75% training and 25% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
# Create the models
# Create the Decision Tree Regressor Model
tree = DecisionTreeRegressor().fit(X_train, y_train)

# Create the Linear Regression Model
lr = LinearRegression().fit(X_train, y_train)
```

```python
# Get the last x rows/days of the Feature Dataset
X_future = df.drop(['Prediction'], 1)[:-future_days]
# I want the last x days
X_future = X_future.tail(future_days)
# I will convert the Dataset to a numpy array
X_future = np.array(X_future)
X_future
```

```python
# Show the model Tree Prediction
tree_prediction = tree.predict(X_future)
print(tree_prediction)
print()
# Show the model Linear Regression Prediction
lr_prediction = lr.predict(X_future)
print(lr_prediction)
```

```python
# Visualize the data

# The predictions
predictions = tree_prediction

# Our valid data
valid = df[X.shape[0]:]
valid['Predictions'] = predictions
plt.figure(figsize = (16, 8))
plt.title('Model')
plt.xlabel('Days')
plt.ylabel('Close Price USD ($)')
plt.plot(df['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Original Data', 'Valid Data', 'Predicted Data'])
plt.show()
```

```python
# The predictions
predictions = lr_prediction

# Our valid data
valid = df[X.shape[0]:]
valid['Predictions'] = predictions
plt.figure(figsize = (16, 8))
plt.title('Model')
plt.xlabel('Days')
plt.ylabel('Close Price USD ($)')
plt.plot(df['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Original Data', 'Valid Data', 'Predicted Data'])
plt.show()
```
