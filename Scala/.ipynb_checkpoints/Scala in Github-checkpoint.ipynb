# The Github history of Scala

```python
# Import Pandas and files
import pandas as pd

pulls = pd.read_csv('pulls.csv')
pull_files = pd.read_csv('pull_files.csv')

pulls.head()
```

### Preparing and cleaning the data

```python
# Convert the Date of the Pulls dataset and convert all times to UTC
pulls['date'] = pd.to_datetime(pulls['date'], utc = True)

pulls.head()
```

### Merging the datasets

```python
# Merge the 2 datasets
data = pulls.merge(pull_files, on = 'pid')

data.head()
```

### Is the project still actively maintained

```python
%matplotlib inline

# Create a column that will store the month and the year, as a string
data['month_year'] = data.apply(lambda x: str(x['date'].year) + '-' + str(x['date'].month), axis = 1)

data.head()
```

### Count the pull requests grouped by month_year

```python
counts = data.groupby('month_year').agg({'month_year': 'count'})

# Plot the results as a barplot
counts.plot.bar(figsize= (50,50))
```

### Is there camaraderie in the project?

```python
%matplotlib inline

# Group by the user
by_user = data.groupby('user').agg({'user': 'count'})

# Plot a histogram of how many users we have
by_user.plot.hist()
```

### What files were changed in the last ten pull requests?

```python
# Identify the last 10 pull requests -- we have 10 entries
last_10 = pulls.nlargest(10, 'date')

# Join the new data set with the files data set -- We have now 34 entries
joined_pr = last_10.merge(pull_files, on = 'pid')

# Identify the unique files
files = set(joined_pr['file'])

# Files changed in the last 10 requests -- The last file modified is Calculate.scala
files
```

### Who made the most pull requests to a given file (the last file modified) ?

```python
# This is the file we are interested in:
file = 'src/compiler/scala/reflect/reify/phases/Calculate.scala'

# Identify the commits that changed the file
file_pr = data[data['file'] == file]

# Count the number of changes made by each user
user_counts = file_pr.groupby('user').agg({'pid': 'count'})

# Print the top 3 users
print(user_counts.nlargest(3, 'pid'))
```

### Who made the last ten pull requests on a given file (the last file)?

```python
# Last file modified
file = 'src/compiler/scala/reflect/reify/phases/Calculate.scala'

# Select the pull requests that changed the last file
file_pr = pull_files[pull_files['file'] == file]

# Merge the obtained results with the pulls DataFrame
joined_pr = file_pr.merge(pulls, on = 'pid')

# Find the users of the last 10 most recent pull requests
users_last_10 = set(joined_pr.nlargest(10, 'date')['user'])

users_last_10
```

### The pull requests of two special developers

```python
%matplotlib inline

# The users we are interested in
authors = ['xeno-by', 'soc']

# Get all the pull requests for these users
by_author = pulls[pulls['user'].isin(authors)]

# Count the number of pull requests submitted each year
counts = by_author.groupby(['user', by_author['date'].dt.year]).agg({'pid': 'count'}).reset_index()

# Convert the table to a wide format
counts_wide = counts.pivot_table(index = 'date', columns = 'user', values = 'pid', fill_value = 0)

# Plot a barplot of the pivot table
counts_wide.plot.bar()
```

### Visualizing the contributions of each developer

```python
# File and users of our interest
authors = ['xeno-by', 'soc']
file = 'src/compiler/scala/reflect/reify/phases/Calculate.scala'

# Select the pull requests submitted by the authors, from the data DataFrame
by_author = data[data['user'].isin(authors)]

# Select the pull requests that affect the file
by_file = by_author[by_author['file'] == file]

# Group and count the number of PRs done by each user each year
grouped = by_file.groupby(['user', by_file['date'].dt.year]).agg({'pid': 'count'}).reset_index()

# Transform the data into a wide format
by_file_wide = grouped.pivot_table(index = 'date', columns = 'user', values = 'pid', fill_value = 0)

# Plot a barplot of the 2 users contributions for the file by year
by_file_wide.plot.bar()
```
