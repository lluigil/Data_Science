# 2. Lire les jeux de donnes

```python
import pandas as pd
# Nous allons lire le jeu de donnes concernant uniquement les couleurs
colors = pd.read_csv('Documents/Lego/colors.csv')
sets = pd.read_csv('Documents/Lego/sets.csv')
```

# 3. Explorer les jeux de donnes

```python
# Combien de couleurs a le jeux de donnes concernant les couleurs
num_colors = len(colors)
num_colors
```

```python
# Rsumer les couleurs : distribution des couleurs par transparence
colors_summary = colors.groupby('is_trans').count()
colors_summary
```

```python
%matplotlib inline
# Rsumer la moyenne du nombre de parties par an
parts_by_year = sets[['year', 'num_parts']].groupby('year', as_index = False).mean()
parts_by_year.plot(x = 'year', y = 'num_parts')
```

```python
# Les blocks Lego sont livrs par thmes, nous allons explorer comment la dlivrance de thmes ont vari par anne
themes_by_year = sets[['year', 'theme_id']].groupby('year', as_index = False).agg({'theme_id': pd.Series.count})
themes_by_year.plot(x = 'year', y = 'theme_id')
```
