# 2. Lire les jeux de données

```python
import pandas as pd
# Nous allons lire le jeu de données concernant uniquement les couleurs
colors = pd.read_csv('Documents/Lego/colors.csv')
sets = pd.read_csv('Documents/Lego/sets.csv')
```

# 3. Explorer les jeux de données

```python
# Combien de couleurs a le jeux de données concernant les couleurs
num_colors = len(colors)
num_colors
```

```python
# Résumer les couleurs : distribution des couleurs par transparence
colors_summary = colors.groupby('is_trans').count()
colors_summary
```

```python
%matplotlib inline
# Résumer la moyenne du nombre de parties par an
parts_by_year = sets[['year', 'num_parts']].groupby('year', as_index = False).mean()
parts_by_year.plot(x = 'year', y = 'num_parts')
```

```python
# Les blocks Lego sont livrés par thèmes, nous allons explorer comment la délivrance de thèmes ont varié par année
themes_by_year = sets[['year', 'theme_id']].groupby('year', as_index = False).agg({'theme_id': pd.Series.count})
themes_by_year.plot(x = 'year', y = 'theme_id')
```
